```python
# ============================================================
# GOOGLE COLAB: ENSEMBLE AI/ML MODEL FOR COGNITIVE OFFLOADING
# FIXED VERSION FOR DENSE MATRIX COMPATIBILITY
# ============================================================

# -----------------------------
# 1. Install packages
# -----------------------------
!pip -q install pandas numpy scikit-learn matplotlib gradio openpyxl

# -----------------------------
# 2. Imports
# -----------------------------
import io
import json
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import (
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    GradientBoostingClassifier,
    HistGradientBoostingClassifier,
    VotingClassifier,
    StackingClassifier,
    RandomForestRegressor,
    ExtraTreesRegressor,
    GradientBoostingRegressor,
    HistGradientBoostingRegressor,
    VotingRegressor,
    StackingRegressor
)
from sklearn.inspection import permutation_importance

import gradio as gr

# -----------------------------
# 3. Upload dataset
# -----------------------------
print("Upload the cognitive offloading CSV file.")
uploaded = files.upload()

if len(uploaded) == 0:
    raise ValueError("No file uploaded.")

filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]))

print(f"\nLoaded file: {filename}")
print(f"Dataset shape: {df.shape}")
display(df.head())

# -----------------------------
# 4. Basic profiling
# -----------------------------
print("\nColumns:")
print(df.columns.tolist())

print("\nMissing values:")
display(df.isna().sum().sort_values(ascending=False).head(30))

# -----------------------------
# 5. Direct cognitive offloading items
# -----------------------------
offloading_items = [
    "rely_ai_suggestions_decisions",
    "follow_ai_over_instinct",
    "less_confident_without_ai_input",
    "ai_decides_detached_from_outcome",
    "question_if_decision_truly_mine_ai",
    "ai_blurs_judgment_automation",
    "worry_ai_weakens_independent_decisions"
]

missing_required = [c for c in offloading_items if c not in df.columns]
if missing_required:
    raise ValueError(f"Missing required columns: {missing_required}")

for col in offloading_items:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# -----------------------------
# 6. Build targets
# -----------------------------
df["Cognitive_Offloading_Index"] = df[offloading_items].mean(axis=1)

q1 = df["Cognitive_Offloading_Index"].quantile(0.33)
q2 = df["Cognitive_Offloading_Index"].quantile(0.67)

def label_offloading(x):
    if pd.isna(x):
        return np.nan
    if x <= q1:
        return "Low"
    elif x <= q2:
        return "Moderate"
    return "High"

df["Cognitive_Offloading_Level"] = df["Cognitive_Offloading_Index"].apply(label_offloading)

print("\nCognitive Offloading Level counts:")
print(df["Cognitive_Offloading_Level"].value_counts(dropna=False))

# -----------------------------
# 7. Feature selection
# -----------------------------
exclude_cols = set(offloading_items + ["Cognitive_Offloading_Index", "Cognitive_Offloading_Level"])
X = df.drop(columns=[c for c in exclude_cols if c in df.columns], errors="ignore").copy()
y_reg = df["Cognitive_Offloading_Index"].copy()
y_clf = df["Cognitive_Offloading_Level"].copy()

# Drop very sparse columns
X = X[[c for c in X.columns if X[c].isna().mean() < 0.60]]

# Convert obvious numeric columns where possible
for col in X.columns:
    if X[col].dtype == "object":
        converted = pd.to_numeric(X[col], errors="coerce")
        # keep conversion only if it preserves enough values
        if converted.notna().sum() > 0.8 * X[col].notna().sum() and converted.notna().sum() > 0:
            X[col] = converted

numeric_features = [c for c in X.columns if pd.api.types.is_numeric_dtype(X[c])]
categorical_features = [c for c in X.columns if c not in numeric_features]

print("\nPredictor columns used:")
print(X.columns.tolist())

print(f"\nNumeric predictors: {len(numeric_features)}")
print(f"Categorical predictors: {len(categorical_features)}")

# -----------------------------
# 8. Preprocessing
# FIX: Force dense output
# -----------------------------
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ],
    sparse_threshold=0
)

# -----------------------------
# 9. Train/test splits
# -----------------------------
clf_df = pd.concat([X, y_clf], axis=1).dropna(subset=["Cognitive_Offloading_Level"]).copy()
X_clf = clf_df[X.columns]
y_clf_clean = clf_df["Cognitive_Offloading_Level"]

Xc_train, Xc_test, yc_train, yc_test = train_test_split(
    X_clf,
    y_clf_clean,
    test_size=0.20,
    random_state=42,
    stratify=y_clf_clean
)

reg_df = pd.concat([X, y_reg], axis=1).dropna(subset=["Cognitive_Offloading_Index"]).copy()
X_reg = reg_df[X.columns]
y_reg_clean = reg_df["Cognitive_Offloading_Index"]

Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_reg,
    y_reg_clean,
    test_size=0.20,
    random_state=42
)

# -----------------------------
# 10. Classification models
# -----------------------------
rf_clf = RandomForestClassifier(n_estimators=300, random_state=42, class_weight="balanced")
et_clf = ExtraTreesClassifier(n_estimators=300, random_state=42, class_weight="balanced")
gb_clf = GradientBoostingClassifier(random_state=42)
hgb_clf = HistGradientBoostingClassifier(random_state=42)

voting_clf = VotingClassifier(
    estimators=[
        ("rf", rf_clf),
        ("et", et_clf),
        ("gb", gb_clf)
    ],
    voting="soft"
)

stacking_clf = StackingClassifier(
    estimators=[
        ("rf", rf_clf),
        ("et", et_clf),
        ("gb", gb_clf),
        ("hgb", hgb_clf)
    ],
    final_estimator=LogisticRegression(max_iter=3000),
    passthrough=False
)

clf_models = {
    "RandomForestClassifier": rf_clf,
    "ExtraTreesClassifier": et_clf,
    "GradientBoostingClassifier": gb_clf,
    "HistGradientBoostingClassifier": hgb_clf,
    "VotingClassifier": voting_clf,
    "StackingClassifier": stacking_clf
}

# -----------------------------
# 11. Regression models
# -----------------------------
rf_reg = RandomForestRegressor(n_estimators=300, random_state=42)
et_reg = ExtraTreesRegressor(n_estimators=300, random_state=42)
gb_reg = GradientBoostingRegressor(random_state=42)
hgb_reg = HistGradientBoostingRegressor(random_state=42)

voting_reg = VotingRegressor(
    estimators=[
        ("rf", rf_reg),
        ("et", et_reg),
        ("gb", gb_reg)
    ]
)

stacking_reg = StackingRegressor(
    estimators=[
        ("rf", rf_reg),
        ("et", et_reg),
        ("gb", gb_reg),
        ("hgb", hgb_reg)
    ],
    final_estimator=Ridge()
)

reg_models = {
    "RandomForestRegressor": rf_reg,
    "ExtraTreesRegressor": et_reg,
    "GradientBoostingRegressor": gb_reg,
    "HistGradientBoostingRegressor": hgb_reg,
    "VotingRegressor": voting_reg,
    "StackingRegressor": stacking_reg
}

# -----------------------------
# 12. Train and evaluate classification models
# -----------------------------
print("\n================ CLASSIFICATION RESULTS ================\n")

best_clf_name = None
best_clf_f1 = -999
best_clf_pipeline = None
clf_results = []

for name, model in clf_models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    pipe.fit(Xc_train, yc_train)
    pred = pipe.predict(Xc_test)

    acc = accuracy_score(yc_test, pred)
    prec = precision_score(yc_test, pred, average="weighted", zero_division=0)
    rec = recall_score(yc_test, pred, average="weighted", zero_division=0)
    f1 = f1_score(yc_test, pred, average="weighted")

    clf_results.append({
        "Model": name,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "Weighted_F1": f1
    })

    print(name)
    print(f"Accuracy:     {acc:.4f}")
    print(f"Precision:    {prec:.4f}")
    print(f"Recall:       {rec:.4f}")
    print(f"Weighted F1:  {f1:.4f}")
    print(classification_report(yc_test, pred))
    print("-" * 70)

    if f1 > best_clf_f1:
        best_clf_f1 = f1
        best_clf_name = name
        best_clf_pipeline = pipe

clf_results_df = pd.DataFrame(clf_results).sort_values("Weighted_F1", ascending=False)
print("\nBest classification model:", best_clf_name)
display(clf_results_df)

# -----------------------------
# 13. Train and evaluate regression models
# -----------------------------
print("\n================ REGRESSION RESULTS ================\n")

best_reg_name = None
best_reg_r2 = -999
best_reg_pipeline = None
reg_results = []

for name, model in reg_models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    pipe.fit(Xr_train, yr_train)
    pred = pipe.predict(Xr_test)

    mae = mean_absolute_error(yr_test, pred)
    rmse = np.sqrt(mean_squared_error(yr_test, pred))
    r2 = r2_score(yr_test, pred)

    reg_results.append({
        "Model": name,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2
    })

    print(name)
    print(f"MAE:   {mae:.4f}")
    print(f"RMSE:  {rmse:.4f}")
    print(f"R2:    {r2:.4f}")
    print("-" * 70)

    if r2 > best_reg_r2:
        best_reg_r2 = r2
        best_reg_name = name
        best_reg_pipeline = pipe

reg_results_df = pd.DataFrame(reg_results).sort_values("R2", ascending=False)
print("\nBest regression model:", best_reg_name)
display(reg_results_df)

# -----------------------------
# 14. Permutation importance
# -----------------------------
print("\n================ FEATURE IMPORTANCE ================\n")

perm_df = None
try:
    perm = permutation_importance(
        best_reg_pipeline,
        Xr_test,
        yr_test,
        n_repeats=10,
        random_state=42,
        scoring="r2"
    )

    perm_df = pd.DataFrame({
        "Feature": X_reg.columns,
        "Importance": perm.importances_mean
    }).sort_values("Importance", ascending=False)

    display(perm_df.head(20))

    plt.figure(figsize=(10, 6))
    topn = min(15, len(perm_df))
    plt.barh(
        perm_df["Feature"].head(topn)[::-1],
        perm_df["Importance"].head(topn)[::-1]
    )
    plt.title("Top Features Predicting Cognitive Offloading")
    plt.xlabel("Permutation Importance")
    plt.tight_layout()
    plt.show()

except Exception as e:
    print("Could not compute permutation importance:", e)

# -----------------------------
# 15. Gradio interface
# -----------------------------
if perm_df is not None and len(perm_df) > 0:
    ui_features = perm_df["Feature"].head(8).tolist()
else:
    ui_features = list(X.columns[:8])

ui_features = [f for f in ui_features if f in X.columns]
ui_features = list(dict.fromkeys(ui_features))

category_options = {}
for col in categorical_features:
    vals = df[col].dropna().astype(str).unique().tolist()
    vals = sorted(vals)[:50]
    category_options[col] = vals

def default_value(col):
    if col in numeric_features:
        s = pd.to_numeric(df[col], errors="coerce")
        return float(s.median()) if s.notna().sum() > 0 else 0.0
    mode_vals = df[col].dropna().astype(str).mode()
    return mode_vals.iloc[0] if len(mode_vals) > 0 else "Unknown"

def predict_cognitive_offloading(*inputs):
    row = {}
    for col, val in zip(ui_features, inputs):
        row[col] = val

    for col in X.columns:
        if col not in row:
            row[col] = default_value(col)

    user_df = pd.DataFrame([row])[X.columns]

    for col in numeric_features:
        user_df[col] = pd.to_numeric(user_df[col], errors="coerce")

    pred_index = float(best_reg_pipeline.predict(user_df)[0])
    pred_level = best_clf_pipeline.predict(user_df)[0]

    try:
        probs = best_clf_pipeline.predict_proba(user_df)[0]
        classes = best_clf_pipeline.named_steps["model"].classes_
        prob_map = {str(cls): round(float(p), 4) for cls, p in zip(classes, probs)}
    except Exception:
        prob_map = {"note": "Probability output unavailable for this model."}

    if pred_index <= q1:
        interpretation = "Predicted cognitive offloading is relatively low."
    elif pred_index <= q2:
        interpretation = "Predicted cognitive offloading is moderate."
    else:
        interpretation = "Predicted cognitive offloading is relatively high."

    result = {
        "Predicted_Cognitive_Offloading_Index": round(pred_index, 4),
        "Predicted_Cognitive_Offloading_Level": str(pred_level),
        "Classification_Probabilities": prob_map,
        "Best_Classification_Model": best_clf_name,
        "Best_Regression_Model": best_reg_name,
        "Interpretation": interpretation
    }

    return json.dumps(result, indent=2)

gradio_inputs = []
for col in ui_features:
    if col in numeric_features:
        s = pd.to_numeric(df[col], errors="coerce")
        cmin = float(s.min()) if s.notna().sum() > 0 else 0.0
        cmax = float(s.max()) if s.notna().sum() > 0 else 10.0
        cmed = float(s.median()) if s.notna().sum() > 0 else 5.0
        step = 1 if abs(cmax - cmin) <= 20 else 0.1

        gradio_inputs.append(
            gr.Slider(
                minimum=cmin,
                maximum=cmax,
                value=cmed,
                step=step,
                label=col
            )
        )
    else:
        vals = category_options.get(col, ["Unknown"])
        default = default_value(col)
        if default not in vals:
            vals = [default] + vals

        gradio_inputs.append(
            gr.Dropdown(
                choices=vals,
                value=default,
                label=col
            )
        )

demo = gr.Interface(
    fn=predict_cognitive_offloading,
    inputs=gradio_inputs,
    outputs=gr.Code(label="Prediction Output", language="json"),
    title="Ensemble AI/ML App for Cognitive Offloading",
    description="This app uses ensemble learning to estimate Cognitive Offloading Index and Cognitive Offloading Level."
)

print("\nLaunching Gradio app...")
demo.launch(debug=True, share=True)
```
