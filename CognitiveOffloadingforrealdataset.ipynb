```python
# ============================================================
# GOOGLE COLAB EXECUTABLE SCRIPT
# XGBOOST + CROSS-VALIDATION + HYPERPARAMETER TUNING
# FOR COGNITIVE OFFLOADING
# ============================================================

# -----------------------------
# 1. Install required packages
# -----------------------------
!pip -q install pandas numpy scikit-learn matplotlib xgboost gradio openpyxl

# -----------------------------
# 2. Imports
# -----------------------------
import io
import json
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files

from sklearn.model_selection import (
    train_test_split,
    StratifiedKFold,
    KFold,
    RandomizedSearchCV,
    cross_val_score
)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    f1_score,
    precision_score,
    recall_score,
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.inspection import permutation_importance
from xgboost import XGBClassifier, XGBRegressor

import gradio as gr

# -----------------------------
# 3. Upload and load dataset
# -----------------------------
print("Upload the cognitive offloading CSV file.")
uploaded = files.upload()

if len(uploaded) == 0:
    raise ValueError("No file uploaded.")

filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]))

print(f"Loaded file: {filename}")
print(f"Dataset shape: {df.shape}")
display(df.head())

# -----------------------------
# 4. Inspect columns and missingness
# -----------------------------
print("Columns:")
print(df.columns.tolist())

print("\nMissing values:")
display(df.isna().sum().sort_values(ascending=False).head(30))

# -----------------------------
# 5. Define direct cognitive offloading indicators
# -----------------------------
offloading_items = [
    "rely_ai_suggestions_decisions",
    "follow_ai_over_instinct",
    "less_confident_without_ai_input",
    "ai_decides_detached_from_outcome",
    "question_if_decision_truly_mine_ai",
    "ai_blurs_judgment_automation",
    "worry_ai_weakens_independent_decisions"
]

missing_required = [c for c in offloading_items if c not in df.columns]
if missing_required:
    raise ValueError(f"Missing required columns: {missing_required}")

for col in offloading_items:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# -----------------------------
# 6. Construct target variables
# -----------------------------
df["Cognitive_Offloading_Index"] = df[offloading_items].mean(axis=1)

q1 = df["Cognitive_Offloading_Index"].quantile(0.33)
q2 = df["Cognitive_Offloading_Index"].quantile(0.67)

def label_offloading(x):
    if pd.isna(x):
        return np.nan
    if x <= q1:
        return "Low"
    elif x <= q2:
        return "Moderate"
    return "High"

df["Cognitive_Offloading_Level"] = df["Cognitive_Offloading_Index"].apply(label_offloading)

print("Cognitive Offloading Level counts:")
print(df["Cognitive_Offloading_Level"].value_counts(dropna=False))

display(df[offloading_items + ["Cognitive_Offloading_Index"]].describe().T)

# -----------------------------
# 7. Reliability check
# -----------------------------
def cronbach_alpha(df_items):
    df_items = df_items.dropna()
    if df_items.shape[1] < 2 or df_items.shape[0] < 2:
        return np.nan
    item_vars = df_items.var(axis=0, ddof=1)
    total_var = df_items.sum(axis=1).var(ddof=1)
    n_items = df_items.shape[1]
    if total_var == 0:
        return np.nan
    alpha = (n_items / (n_items - 1)) * (1 - item_vars.sum() / total_var)
    return alpha

alpha = cronbach_alpha(df[offloading_items])
print(f"Cronbach's alpha: {alpha:.3f}")

# -----------------------------
# 8. Prepare predictor variables
# -----------------------------
exclude_cols = set(offloading_items + ["Cognitive_Offloading_Index", "Cognitive_Offloading_Level"])
X = df.drop(columns=[c for c in exclude_cols if c in df.columns], errors="ignore").copy()
y_reg = df["Cognitive_Offloading_Index"].copy()
y_clf = df["Cognitive_Offloading_Level"].copy()

# Remove columns with excessive missingness
X = X[[c for c in X.columns if X[c].isna().mean() < 0.60]]

# Convert object columns to numeric where sensible
for col in X.columns:
    if X[col].dtype == "object":
        converted = pd.to_numeric(X[col], errors="coerce")
        original_non_null = X[col].notna().sum()
        converted_non_null = converted.notna().sum()
        if original_non_null > 0 and converted_non_null / original_non_null >= 0.80:
            X[col] = converted

numeric_features = [c for c in X.columns if pd.api.types.is_numeric_dtype(X[c])]
categorical_features = [c for c in X.columns if c not in numeric_features]

print("Predictor columns used:")
print(X.columns.tolist())
print(f"\nNumeric predictors: {len(numeric_features)}")
print(f"Categorical predictors: {len(categorical_features)}")

# -----------------------------
# 9. Preprocessing pipeline
# -----------------------------
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ],
    sparse_threshold=0
)

# -----------------------------
# 10. Split the data
# -----------------------------
# Classification
clf_df = pd.concat([X, y_clf], axis=1).dropna(subset=["Cognitive_Offloading_Level"]).copy()
X_clf = clf_df[X.columns]
y_clf_clean = clf_df["Cognitive_Offloading_Level"]

label_encoder = LabelEncoder()
y_clf_encoded = label_encoder.fit_transform(y_clf_clean)

Xc_train, Xc_test, yc_train, yc_test = train_test_split(
    X_clf,
    y_clf_encoded,
    test_size=0.20,
    random_state=42,
    stratify=y_clf_encoded
)

# Regression
reg_df = pd.concat([X, y_reg], axis=1).dropna(subset=["Cognitive_Offloading_Index"]).copy()
X_reg = reg_df[X.columns]
y_reg_clean = reg_df["Cognitive_Offloading_Index"]

Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_reg,
    y_reg_clean,
    test_size=0.20,
    random_state=42
)

# -----------------------------
# 11. Define baseline models
# -----------------------------
baseline_clf = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(
        n_estimators=300,
        random_state=42,
        class_weight="balanced"
    ))
])

baseline_reg = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestRegressor(
        n_estimators=300,
        random_state=42
    ))
])

# -----------------------------
# 12. Define XGBoost models
# -----------------------------
xgb_clf_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", XGBClassifier(
        objective="multi:softprob",
        eval_metric="mlogloss",
        random_state=42,
        n_jobs=-1
    ))
])

xgb_reg_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", XGBRegressor(
        objective="reg:squarederror",
        eval_metric="rmse",
        random_state=42,
        n_jobs=-1
    ))
])

clf_param_dist = {
    "model__n_estimators": [100, 200, 300, 500],
    "model__max_depth": [3, 4, 5, 6, 8],
    "model__learning_rate": [0.01, 0.03, 0.05, 0.1, 0.2],
    "model__subsample": [0.7, 0.8, 0.9, 1.0],
    "model__colsample_bytree": [0.6, 0.7, 0.8, 0.9, 1.0],
    "model__min_child_weight": [1, 3, 5, 7],
    "model__gamma": [0, 0.1, 0.3, 0.5]
}

reg_param_dist = {
    "model__n_estimators": [100, 200, 300, 500],
    "model__max_depth": [3, 4, 5, 6, 8],
    "model__learning_rate": [0.01, 0.03, 0.05, 0.1, 0.2],
    "model__subsample": [0.7, 0.8, 0.9, 1.0],
    "model__colsample_bytree": [0.6, 0.7, 0.8, 0.9, 1.0],
    "model__min_child_weight": [1, 3, 5, 7],
    "model__gamma": [0, 0.1, 0.3, 0.5]
}

# -----------------------------
# 13. Tune classifier with cross-validation
# -----------------------------
cv_clf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

xgb_clf_search = RandomizedSearchCV(
    estimator=xgb_clf_pipeline,
    param_distributions=clf_param_dist,
    n_iter=25,
    scoring="f1_weighted",
    cv=cv_clf,
    verbose=1,
    n_jobs=-1,
    random_state=42
)

print("Fitting tuned XGBoost classifier...")
xgb_clf_search.fit(Xc_train, yc_train)

best_xgb_clf = xgb_clf_search.best_estimator_

baseline_clf_scores = cross_val_score(
    baseline_clf,
    Xc_train,
    yc_train,
    cv=cv_clf,
    scoring="f1_weighted",
    n_jobs=-1
)

xgb_clf_scores = cross_val_score(
    best_xgb_clf,
    Xc_train,
    yc_train,
    cv=cv_clf,
    scoring="f1_weighted",
    n_jobs=-1
)

print("Baseline RF CV Weighted F1:", baseline_clf_scores.mean(), "+/-", baseline_clf_scores.std())
print("Tuned XGBoost CV Weighted F1:", xgb_clf_scores.mean(), "+/-", xgb_clf_scores.std())
print("Best classification parameters:")
print(xgb_clf_search.best_params_)

# -----------------------------
# 14. Evaluate classification models
# -----------------------------
baseline_clf.fit(Xc_train, yc_train)
baseline_clf_pred = baseline_clf.predict(Xc_test)

best_xgb_clf.fit(Xc_train, yc_train)
xgb_clf_pred = best_xgb_clf.predict(Xc_test)

clf_comparison = []
for model_name, y_pred in [
    ("Baseline Random Forest", baseline_clf_pred),
    ("Tuned XGBoost", xgb_clf_pred)
]:
    clf_comparison.append({
        "Model": model_name,
        "Accuracy": accuracy_score(yc_test, y_pred),
        "Precision": precision_score(yc_test, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(yc_test, y_pred, average="weighted", zero_division=0),
        "Weighted_F1": f1_score(yc_test, y_pred, average="weighted")
    })

clf_comparison_df = pd.DataFrame(clf_comparison).sort_values("Weighted_F1", ascending=False)
display(clf_comparison_df)

best_clf_name = clf_comparison_df.iloc[0]["Model"]
best_clf_pipeline = best_xgb_clf if best_clf_name == "Tuned XGBoost" else baseline_clf
best_clf_pred = xgb_clf_pred if best_clf_name == "Tuned XGBoost" else baseline_clf_pred

print("Best classification model:", best_clf_name)
print(classification_report(
    yc_test,
    best_clf_pred,
    target_names=label_encoder.classes_
))
print("Confusion matrix:")
print(confusion_matrix(yc_test, best_clf_pred))

# -----------------------------
# 15. Tune regressor with cross-validation
# -----------------------------
cv_reg = KFold(n_splits=5, shuffle=True, random_state=42)

xgb_reg_search = RandomizedSearchCV(
    estimator=xgb_reg_pipeline,
    param_distributions=reg_param_dist,
    n_iter=25,
    scoring="r2",
    cv=cv_reg,
    verbose=1,
    n_jobs=-1,
    random_state=42
)

print("Fitting tuned XGBoost regressor...")
xgb_reg_search.fit(Xr_train, yr_train)

best_xgb_reg = xgb_reg_search.best_estimator_

baseline_reg_scores = cross_val_score(
    baseline_reg,
    Xr_train,
    yr_train,
    cv=cv_reg,
    scoring="r2",
    n_jobs=-1
)

xgb_reg_scores = cross_val_score(
    best_xgb_reg,
    Xr_train,
    yr_train,
    cv=cv_reg,
    scoring="r2",
    n_jobs=-1
)

print("Baseline RF CV R2:", baseline_reg_scores.mean(), "+/-", baseline_reg_scores.std())
print("Tuned XGBoost CV R2:", xgb_reg_scores.mean(), "+/-", xgb_reg_scores.std())
print("Best regression parameters:")
print(xgb_reg_search.best_params_)

# -----------------------------
# 16. Evaluate regression models
# -----------------------------
baseline_reg.fit(Xr_train, yr_train)
baseline_reg_pred = baseline_reg.predict(Xr_test)

best_xgb_reg.fit(Xr_train, yr_train)
xgb_reg_pred = best_xgb_reg.predict(Xr_test)

reg_comparison = []
for model_name, y_pred in [
    ("Baseline Random Forest", baseline_reg_pred),
    ("Tuned XGBoost", xgb_reg_pred)
]:
    reg_comparison.append({
        "Model": model_name,
        "MAE": mean_absolute_error(yr_test, y_pred),
        "RMSE": np.sqrt(mean_squared_error(yr_test, y_pred)),
        "R2": r2_score(yr_test, y_pred)
    })

reg_comparison_df = pd.DataFrame(reg_comparison).sort_values("R2", ascending=False)
display(reg_comparison_df)

best_reg_name = reg_comparison_df.iloc[0]["Model"]
best_reg_pipeline = best_xgb_reg if best_reg_name == "Tuned XGBoost" else baseline_reg

print("Best regression model:", best_reg_name)

# -----------------------------
# 17. Permutation importance
# -----------------------------
perm = permutation_importance(
    best_reg_pipeline,
    Xr_test,
    yr_test,
    n_repeats=10,
    random_state=42,
    scoring="r2"
)

perm_df = pd.DataFrame({
    "Feature": X_reg.columns,
    "Importance": perm.importances_mean
}).sort_values("Importance", ascending=False)

display(perm_df.head(20))

plt.figure(figsize=(10, 6))
topn = min(15, len(perm_df))
plt.barh(
    perm_df["Feature"].head(topn)[::-1],
    perm_df["Importance"].head(topn)[::-1]
)
plt.title("Top Features Predicting Cognitive Offloading")
plt.xlabel("Permutation Importance")
plt.tight_layout()
plt.show()

# -----------------------------
# 18. Summary interpretation
# -----------------------------
print("===== MODEL SUMMARY =====")
print(f"Best classification model: {best_clf_name}")
print(f"Best regression model: {best_reg_name}")
print()
print("Classification comparison:")
display(clf_comparison_df)
print()
print("Regression comparison:")
display(reg_comparison_df)

# -----------------------------
# 19. Build the Gradio interface
# -----------------------------
if len(perm_df) > 0:
    ui_features = perm_df["Feature"].head(8).tolist()
else:
    ui_features = list(X.columns[:8])

ui_features = [f for f in ui_features if f in X.columns]
ui_features = list(dict.fromkeys(ui_features))

category_options = {}
for col in categorical_features:
    vals = df[col].dropna().astype(str).unique().tolist()
    vals = sorted(vals)[:50]
    category_options[col] = vals

def default_value(col):
    if col in numeric_features:
        s = pd.to_numeric(df[col], errors="coerce")
        return float(s.median()) if s.notna().sum() > 0 else 0.0
    mode_vals = df[col].dropna().astype(str).mode()
    return mode_vals.iloc[0] if len(mode_vals) > 0 else "Unknown"

def predict_cognitive_offloading(*inputs):
    row = {}
    for col, val in zip(ui_features, inputs):
        row[col] = val

    for col in X.columns:
        if col not in row:
            row[col] = default_value(col)

    user_df = pd.DataFrame([row])[X.columns]

    for col in numeric_features:
        user_df[col] = pd.to_numeric(user_df[col], errors="coerce")

    pred_index = float(best_reg_pipeline.predict(user_df)[0])

    best_clf_pred_encoded = best_clf_pipeline.predict(user_df)[0]
    pred_level = label_encoder.inverse_transform([best_clf_pred_encoded])[0]

    try:
        probs = best_clf_pipeline.predict_proba(user_df)[0]
        classes = label_encoder.inverse_transform(np.arange(len(probs)))
        prob_map = {str(cls): round(float(p), 4) for cls, p in zip(classes, probs)}
    except Exception:
        prob_map = {"note": "Probability output unavailable for this model."}

    if pred_index <= q1:
        interpretation = "Predicted cognitive offloading is relatively low."
    elif pred_index <= q2:
        interpretation = "Predicted cognitive offloading is moderate."
    else:
        interpretation = "Predicted cognitive offloading is relatively high."

    result = {
        "Predicted_Cognitive_Offloading_Index": round(pred_index, 4),
        "Predicted_Cognitive_Offloading_Level": str(pred_level),
        "Classification_Probabilities": prob_map,
        "Best_Classification_Model": best_clf_name,
        "Best_Regression_Model": best_reg_name,
        "Interpretation": interpretation
    }

    return json.dumps(result, indent=2)

gradio_inputs = []
for col in ui_features:
    if col in numeric_features:
        s = pd.to_numeric(df[col], errors="coerce")
        cmin = float(s.min()) if s.notna().sum() > 0 else 0.0
        cmax = float(s.max()) if s.notna().sum() > 0 else 10.0
        cmed = float(s.median()) if s.notna().sum() > 0 else 5.0
        step = 1 if abs(cmax - cmin) <= 20 else 0.1

        gradio_inputs.append(
            gr.Slider(
                minimum=cmin,
                maximum=cmax,
                value=cmed,
                step=step,
                label=col
            )
        )
    else:
        vals = category_options.get(col, ["Unknown"])
        default = default_value(col)
        if default not in vals:
            vals = [default] + vals

        gradio_inputs.append(
            gr.Dropdown(
                choices=vals,
                value=default,
                label=col
            )
        )

demo = gr.Interface(
    fn=predict_cognitive_offloading,
    inputs=gradio_inputs,
    outputs=gr.Code(label="Prediction Output", language="json"),
    title="XGBoost App for Cognitive Offloading",
    description="This app uses tuned XGBoost models with cross-validation to estimate Cognitive Offloading Index and Cognitive Offloading Level."
)

print("Launching Gradio app...")
demo.launch(debug=True, share=True)
```
