```python
# ============================================================
# GOOGLE COLAB: ENSEMBLE AI/ML MODEL TO EXAMINE
# COGNITIVE OFFLOADING IN THE ATTACHED DATASET
# ============================================================

# -----------------------------
# 1. Install packages
# -----------------------------
!pip -q install pandas numpy scikit-learn openpyxl matplotlib gradio

# -----------------------------
# 2. Imports
# -----------------------------
import io
import json
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files

from sklearn.model_selection import train_test_split, StratifiedKFold, KFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    GradientBoostingClassifier,
    HistGradientBoostingClassifier,
    VotingClassifier,
    StackingClassifier,
    RandomForestRegressor,
    ExtraTreesRegressor,
    GradientBoostingRegressor,
    HistGradientBoostingRegressor,
    VotingRegressor,
    StackingRegressor
)
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.inspection import permutation_importance

import gradio as gr

# -----------------------------
# 3. Upload and load dataset
# -----------------------------
print("Upload the attached Excel dataset: synthetic_cognitive_offloading_650_ai_users.xlsx")
uploaded = files.upload()

if len(uploaded) == 0:
    raise ValueError("No file uploaded.")

filename = list(uploaded.keys())[0]

if filename.lower().endswith(".xlsx") or filename.lower().endswith(".xls"):
    df = pd.read_excel(io.BytesIO(uploaded[filename]))
elif filename.lower().endswith(".csv"):
    df = pd.read_csv(io.BytesIO(uploaded[filename]))
else:
    raise ValueError("Unsupported file type. Please upload the Excel or CSV dataset.")

print(f"Loaded file: {filename}")
print(f"Dataset shape: {df.shape}")
display(df.head())

# -----------------------------
# 4. Basic inspection
# -----------------------------
print("Columns:")
print(df.columns.tolist())

print("\nMissing values (top 25):")
display(df.isna().sum().sort_values(ascending=False).head(25))

print("\nData types:")
display(pd.DataFrame({"column": df.columns, "dtype": df.dtypes.astype(str).values}))

# -----------------------------
# 5. Core target definitions
# -----------------------------
# Use existing targets if present; otherwise build from direct offloading items.

direct_offloading_items = [
    "rely_ai_suggestions_decisions",
    "follow_ai_over_instinct",
    "less_confident_without_ai_input",
    "ai_decides_detached_from_outcome",
    "question_if_decision_truly_mine_ai",
    "ai_blurs_judgment_automation",
    "worry_ai_weakens_independent_decisions"
]

for col in direct_offloading_items:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

if "cognitive_offloading_index" not in df.columns:
    missing_items = [c for c in direct_offloading_items if c not in df.columns]
    if missing_items:
        raise ValueError(f"Cannot build cognitive_offloading_index. Missing columns: {missing_items}")
    df["cognitive_offloading_index"] = df[direct_offloading_items].mean(axis=1)

if "cognitive_offloading_level" not in df.columns:
    q1 = df["cognitive_offloading_index"].quantile(0.33)
    q2 = df["cognitive_offloading_index"].quantile(0.67)

    def label_offloading(x):
        if pd.isna(x):
            return np.nan
        if x <= q1:
            return "Low"
        elif x <= q2:
            return "Moderate"
        return "High"

    df["cognitive_offloading_level"] = df["cognitive_offloading_index"].apply(label_offloading)

# Standardize labels
df["cognitive_offloading_level"] = (
    df["cognitive_offloading_level"]
    .astype(str)
    .str.strip()
    .str.title()
)

print("\nCognitive offloading level counts:")
print(df["cognitive_offloading_level"].value_counts(dropna=False))

print("\nDescriptive statistics for selected core fields:")
core_stats_cols = [c for c in [
    "cognitive_offloading_index",
    "ai_use_frequency",
    "verification_rate",
    "ai_output_acceptance_rate",
    "task_accuracy_without_ai",
    "performance_drop_without_ai",
    "trust_in_ai_score",
    "confidence_without_ai"
] if c in df.columns]

if core_stats_cols:
    display(df[core_stats_cols].describe().T)

# -----------------------------
# 6. Reliability check
# -----------------------------
def cronbach_alpha(df_items: pd.DataFrame) -> float:
    df_items = df_items.dropna()
    if df_items.shape[1] < 2 or df_items.shape[0] < 2:
        return np.nan
    item_vars = df_items.var(axis=0, ddof=1)
    total_var = df_items.sum(axis=1).var(ddof=1)
    n_items = df_items.shape[1]
    if total_var == 0:
        return np.nan
    return (n_items / (n_items - 1)) * (1 - item_vars.sum() / total_var)

if all(col in df.columns for col in direct_offloading_items):
    alpha = cronbach_alpha(df[direct_offloading_items])
    print(f"Cronbach's alpha for the seven direct offloading items: {alpha:.3f}")

# -----------------------------
# 7. Prepare predictors
# -----------------------------
# Exclude IDs, timestamps, text notes, and direct target columns from predictors.
exclude_cols = {
    "record_id",
    "participant_id",
    "session_id",
    "timestamp",
    "task_id",
    "ai_use_reflection",
    "decision_process_note",
    "verification_strategy_note",
    "cognitive_offloading_index",
    "cognitive_offloading_level"
}

# Also exclude direct items used to define the target to reduce leakage
for col in direct_offloading_items:
    exclude_cols.add(col)

X = df.drop(columns=[c for c in exclude_cols if c in df.columns], errors="ignore").copy()
y_reg = df["cognitive_offloading_index"].copy()
y_clf = df["cognitive_offloading_level"].copy()

# Drop columns with too much missingness
X = X[[c for c in X.columns if X[c].isna().mean() < 0.60]]

# Try to coerce numeric-like object columns
for col in X.columns:
    if X[col].dtype == "object":
        converted = pd.to_numeric(X[col], errors="coerce")
        non_null_original = X[col].notna().sum()
        non_null_converted = converted.notna().sum()
        if non_null_original > 0 and non_null_converted / non_null_original >= 0.85:
            X[col] = converted

numeric_features = [c for c in X.columns if pd.api.types.is_numeric_dtype(X[c])]
categorical_features = [c for c in X.columns if c not in numeric_features]

print("\nPredictor columns used:")
print(X.columns.tolist())
print(f"\nNumeric predictors: {len(numeric_features)}")
print(f"Categorical predictors: {len(categorical_features)}")

# -----------------------------
# 8. Preprocessing
# -----------------------------
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ],
    sparse_threshold=0
)

# -----------------------------
# 9. Train/test split
# -----------------------------
# Classification
clf_df = pd.concat([X, y_clf], axis=1).dropna(subset=["cognitive_offloading_level"]).copy()
X_clf = clf_df[X.columns]
y_clf_clean = clf_df["cognitive_offloading_level"]

label_encoder = LabelEncoder()
y_clf_encoded = label_encoder.fit_transform(y_clf_clean)

Xc_train, Xc_test, yc_train, yc_test = train_test_split(
    X_clf,
    y_clf_encoded,
    test_size=0.20,
    random_state=42,
    stratify=y_clf_encoded
)

# Regression
reg_df = pd.concat([X, y_reg], axis=1).dropna(subset=["cognitive_offloading_index"]).copy()
X_reg = reg_df[X.columns]
y_reg_clean = reg_df["cognitive_offloading_index"]

Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_reg,
    y_reg_clean,
    test_size=0.20,
    random_state=42
)

# -----------------------------
# 10. Define classification models
# -----------------------------
rf_clf = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced"
)

et_clf = ExtraTreesClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced"
)

gb_clf = GradientBoostingClassifier(random_state=42)
hgb_clf = HistGradientBoostingClassifier(random_state=42)

voting_clf = VotingClassifier(
    estimators=[
        ("rf", rf_clf),
        ("et", et_clf),
        ("gb", gb_clf)
    ],
    voting="soft"
)

stacking_clf = StackingClassifier(
    estimators=[
        ("rf", rf_clf),
        ("et", et_clf),
        ("gb", gb_clf),
        ("hgb", hgb_clf)
    ],
    final_estimator=LogisticRegression(max_iter=3000),
    passthrough=False
)

clf_models = {
    "RandomForestClassifier": rf_clf,
    "ExtraTreesClassifier": et_clf,
    "GradientBoostingClassifier": gb_clf,
    "HistGradientBoostingClassifier": hgb_clf,
    "VotingClassifier": voting_clf,
    "StackingClassifier": stacking_clf
}

# -----------------------------
# 11. Define regression models
# -----------------------------
rf_reg = RandomForestRegressor(n_estimators=300, random_state=42)
et_reg = ExtraTreesRegressor(n_estimators=300, random_state=42)
gb_reg = GradientBoostingRegressor(random_state=42)
hgb_reg = HistGradientBoostingRegressor(random_state=42)

voting_reg = VotingRegressor(
    estimators=[
        ("rf", rf_reg),
        ("et", et_reg),
        ("gb", gb_reg)
    ]
)

stacking_reg = StackingRegressor(
    estimators=[
        ("rf", rf_reg),
        ("et", et_reg),
        ("gb", gb_reg),
        ("hgb", hgb_reg)
    ],
    final_estimator=Ridge()
)

reg_models = {
    "RandomForestRegressor": rf_reg,
    "ExtraTreesRegressor": et_reg,
    "GradientBoostingRegressor": gb_reg,
    "HistGradientBoostingRegressor": hgb_reg,
    "VotingRegressor": voting_reg,
    "StackingRegressor": stacking_reg
}

# -----------------------------
# 12. Cross-validation helper
# -----------------------------
print("\n================ CROSS-VALIDATION ================\n")

cv_clf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_reg = KFold(n_splits=5, shuffle=True, random_state=42)

cv_clf_results = []
for name, model in clf_models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])
    scores = cross_val_score(pipe, Xc_train, yc_train, cv=cv_clf, scoring="f1_weighted", n_jobs=-1)
    cv_clf_results.append({
        "Model": name,
        "CV_Weighted_F1_Mean": scores.mean(),
        "CV_Weighted_F1_SD": scores.std()
    })

cv_clf_results_df = pd.DataFrame(cv_clf_results).sort_values("CV_Weighted_F1_Mean", ascending=False)
display(cv_clf_results_df)

cv_reg_results = []
for name, model in reg_models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])
    scores = cross_val_score(pipe, Xr_train, yr_train, cv=cv_reg, scoring="r2", n_jobs=-1)
    cv_reg_results.append({
        "Model": name,
        "CV_R2_Mean": scores.mean(),
        "CV_R2_SD": scores.std()
    })

cv_reg_results_df = pd.DataFrame(cv_reg_results).sort_values("CV_R2_Mean", ascending=False)
display(cv_reg_results_df)

# -----------------------------
# 13. Train and evaluate classification models
# -----------------------------
print("\n================ CLASSIFICATION RESULTS ================\n")

best_clf_name = None
best_clf_f1 = -999
best_clf_pipeline = None
clf_results = []

for name, model in clf_models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    pipe.fit(Xc_train, yc_train)
    pred = pipe.predict(Xc_test)

    acc = accuracy_score(yc_test, pred)
    prec = precision_score(yc_test, pred, average="weighted", zero_division=0)
    rec = recall_score(yc_test, pred, average="weighted", zero_division=0)
    f1 = f1_score(yc_test, pred, average="weighted")

    clf_results.append({
        "Model": name,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "Weighted_F1": f1
    })

    print(name)
    print(f"Accuracy:     {acc:.4f}")
    print(f"Precision:    {prec:.4f}")
    print(f"Recall:       {rec:.4f}")
    print(f"Weighted F1:  {f1:.4f}")
    print(classification_report(yc_test, pred, target_names=label_encoder.classes_))
    print("-" * 70)

    if f1 > best_clf_f1:
        best_clf_f1 = f1
        best_clf_name = name
        best_clf_pipeline = pipe

clf_results_df = pd.DataFrame(clf_results).sort_values("Weighted_F1", ascending=False)
print("\nBest classification model:", best_clf_name)
display(clf_results_df)

# -----------------------------
# 14. Train and evaluate regression models
# -----------------------------
print("\n================ REGRESSION RESULTS ================\n")

best_reg_name = None
best_reg_r2 = -999
best_reg_pipeline = None
reg_results = []

for name, model in reg_models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    pipe.fit(Xr_train, yr_train)
    pred = pipe.predict(Xr_test)

    mae = mean_absolute_error(yr_test, pred)
    rmse = np.sqrt(mean_squared_error(yr_test, pred))
    r2 = r2_score(yr_test, pred)

    reg_results.append({
        "Model": name,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2
    })

    print(name)
    print(f"MAE:   {mae:.4f}")
    print(f"RMSE:  {rmse:.4f}")
    print(f"R2:    {r2:.4f}")
    print("-" * 70)

    if r2 > best_reg_r2:
        best_reg_r2 = r2
        best_reg_name = name
        best_reg_pipeline = pipe

reg_results_df = pd.DataFrame(reg_results).sort_values("R2", ascending=False)
print("\nBest regression model:", best_reg_name)
display(reg_results_df)

# -----------------------------
# 15. Feature importance
# -----------------------------
print("\n================ FEATURE IMPORTANCE ================\n")

perm_df = None
try:
    perm = permutation_importance(
        best_reg_pipeline,
        Xr_test,
        yr_test,
        n_repeats=10,
        random_state=42,
        scoring="r2"
    )

    perm_df = pd.DataFrame({
        "Feature": X_reg.columns,
        "Importance": perm.importances_mean
    }).sort_values("Importance", ascending=False)

    display(perm_df.head(20))

    plt.figure(figsize=(10, 6))
    topn = min(15, len(perm_df))
    plt.barh(
        perm_df["Feature"].head(topn)[::-1],
        perm_df["Importance"].head(topn)[::-1]
    )
    plt.title("Top Features Predicting Cognitive Offloading")
    plt.xlabel("Permutation Importance")
    plt.tight_layout()
    plt.show()

except Exception as e:
    print("Could not compute permutation importance:", e)

# -----------------------------
# 16. Summary
# -----------------------------
print("\n===== MODEL SUMMARY =====")
print(f"Best classification model: {best_clf_name}")
print(f"Best regression model: {best_reg_name}")
print("\nClassification comparison:")
display(clf_results_df)
print("\nRegression comparison:")
display(reg_results_df)

# -----------------------------
# 17. Build Gradio UI
# -----------------------------
if perm_df is not None and len(perm_df) > 0:
    ui_features = perm_df["Feature"].head(8).tolist()
else:
    ui_features = list(X.columns[:8])

ui_features = [f for f in ui_features if f in X.columns]
ui_features = list(dict.fromkeys(ui_features))

category_options = {}
for col in categorical_features:
    vals = df[col].dropna().astype(str).unique().tolist()
    vals = sorted(vals)[:50]
    category_options[col] = vals

def default_value(col):
    if col in numeric_features:
        s = pd.to_numeric(df[col], errors="coerce")
        return float(s.median()) if s.notna().sum() > 0 else 0.0
    mode_vals = df[col].dropna().astype(str).mode()
    return mode_vals.iloc[0] if len(mode_vals) > 0 else "Unknown"

def predict_cognitive_offloading(*inputs):
    row = {}
    for col, val in zip(ui_features, inputs):
        row[col] = val

    for col in X.columns:
        if col not in row:
            row[col] = default_value(col)

    user_df = pd.DataFrame([row])[X.columns]

    for col in numeric_features:
        user_df[col] = pd.to_numeric(user_df[col], errors="coerce")

    pred_index = float(best_reg_pipeline.predict(user_df)[0])

    best_clf_pred_encoded = best_clf_pipeline.predict(user_df)[0]
    pred_level = label_encoder.inverse_transform([best_clf_pred_encoded])[0]

    try:
        probs = best_clf_pipeline.predict_proba(user_df)[0]
        classes = label_encoder.inverse_transform(np.arange(len(probs)))
        prob_map = {str(cls): round(float(p), 4) for cls, p in zip(classes, probs)}
    except Exception:
        prob_map = {"note": "Probability output unavailable for this model."}

    result = {
        "Predicted_Cognitive_Offloading_Index": round(pred_index, 4),
        "Predicted_Cognitive_Offloading_Level": str(pred_level),
        "Classification_Probabilities": prob_map,
        "Best_Classification_Model": best_clf_name,
        "Best_Regression_Model": best_reg_name
    }

    return json.dumps(result, indent=2)

gradio_inputs = []
for col in ui_features:
    if col in numeric_features:
        s = pd.to_numeric(df[col], errors="coerce")
        cmin = float(s.min()) if s.notna().sum() > 0 else 0.0
        cmax = float(s.max()) if s.notna().sum() > 0 else 10.0
        cmed = float(s.median()) if s.notna().sum() > 0 else 5.0
        step = 1 if abs(cmax - cmin) <= 20 else 0.1

        gradio_inputs.append(
            gr.Slider(
                minimum=cmin,
                maximum=cmax,
                value=cmed,
                step=step,
                label=col
            )
        )
    else:
        vals = category_options.get(col, ["Unknown"])
        default = default_value(col)
        if default not in vals:
            vals = [default] + vals

        gradio_inputs.append(
            gr.Dropdown(
                choices=vals,
                value=default,
                label=col
            )
        )

demo = gr.Interface(
    fn=predict_cognitive_offloading,
    inputs=gradio_inputs,
    outputs=gr.Code(label="Prediction Output", language="json"),
    title="Ensemble AI/ML App for Cognitive Offloading",
    description="This app uses ensemble learning to examine cognitive offloading in the uploaded synthetic dataset."
)

print("Launching Gradio app...")
demo.launch(debug=True, share=True)
```
